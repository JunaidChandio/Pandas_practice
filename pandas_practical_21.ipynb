```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                 'B':[10,20,30]})
df2=pd.DataFrame({'C':[4,5,6],
                 'D':[40,50,60]})

display(df1,df2)
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                 'B':[10,20,30]},index=['a','b','c'])
df2=pd.DataFrame({'C':[4,5,6],
                 'D':[40,50,60]})

display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                 'B':[10,20,30]},index=['a','b','c'])
df2=pd.DataFrame({'C':[4,5,6],
                 'D':[40,50,60]},index=['a','b','c'])

display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                 'B':[10,20,30]},index=['a','b','c'])
df2=pd.DataFrame({'C':[4,5],
                 'D':[40,50]},index=['a','b'])

display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='right')
```

```python
df1.join(df2,how='inner')
```

```python
df1.join(df2,how='outer')
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                 'B':[10,20,30]},index=['a','b','c'])
df2=pd.DataFrame({'A':[4,5],
                 'D':[40,50]},index=['a','b'])

display(df1,df2)
```

```python
df1.join(df2,rsuffix='_right')
```

```python
df1.join(df2,lsuffix='_left')
```
